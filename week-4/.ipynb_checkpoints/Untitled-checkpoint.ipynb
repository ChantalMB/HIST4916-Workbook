```python
# we need this so that we can grab stuff off the web
import requests

# target webpage
url = "https://www.canadiana.ca/view/oocihm.95242/11?r=0&s=1"

# Getting the webpage, creating a Response object.
response = requests.get(url)
 
# Extracting the source code of the page.
data = response.text
 
from bs4 import BeautifulSoup
import lxml

#we give that data to BS so that it can extract what we're interested in
soup = BeautifulSoup(data, 'lxml')

print(soup.prettify())
```

```python
# the 'option' tag is present for each image bc of the dropdown navigation
# find number of images there are in source by counting 'option' tag 
pgCount = soup.find_all('option')
print(len(pgCount))

imgs = soup.find_all('img')

# figure out id of img src
for img in imgs:
    print(img)
    
```

```python
# grab src attribute
srcImg = soup.find_all(id='pvImg')

for url in srcImg:
    print(url['src'])
```
